```python
import torch
a = torch.rand(2,3)
b = torch.rand(2,3)
print(f"a={a}\nb={b}")
```

```python
a_T = a.T
print(f"a_T={a_T}")
```

MAtrix Multiplication

```python
dot_AB = a @ b.T
print(f"dot_AB={dot_AB}")
```

Broadcasting

```python
a = torch.tensor([[1,2,3],[4,5,6]])
b = torch.tensor([[7,8,9]])
print(f"a={a}\nb={b}")
```

```python
c = a + b
print(f"c={c}")
```

Batch MAtrix Multiplication

```python
batch_size = 8
a = torch.rand(batch_size, 2, 3) # a with mtrix of 2x3, batch size of 8 that is 8 matrices of 2x3 - this is a tensor of 3 dimensions
b = torch.rand(batch_size, 3, 4) # b with mtrix of 3x4, batch size of 8 that is 8 matrices of 3x4 - this is a tensor of 3 dimensions
print(f"Shapes: a={a.shape}\nb={b.shape}")
```

```python
c = a@b
print(f"c={c}\nShape of c={c.shape}")
```

Batch tensor multiplication , using bre=oadcasting

```python
d = torch.rand(3,4)
e = a @ d
print(f"e={e}\nShape of e={e.shape}")
```

Reshape

```python
a = torch.randn(size=[2,4, 5])
a_flat = a.reshape(-1)
print(f"a={a}\na_flat={a_flat}")
```
